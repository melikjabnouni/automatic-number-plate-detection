````python
import os
import xml.etree.ElementTree as ET

# Dossier o√π sont stock√©es les images et annotations
train_images_path = "/content/dataset/train"
output_annotations_path = "/content/dataset/train/labels"

# Cr√©er le dossier pour les annotations s'il n'existe pas
os.makedirs(output_annotations_path, exist_ok=True)

# Lister toutes les images dans le dossier
image_files = [f for f in os.listdir(train_images_path) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Fonction pour normaliser les coordonn√©es
def normalize_coordinates(xmin, ymin, xmax, ymax, img_width, img_height):
    center_x = (xmin + xmax) / 2 / img_width
    center_y = (ymin + ymax) / 2 / img_height
    width = (xmax - xmin) / img_width
    height = (ymax - ymin) / img_height
    return center_x, center_y, width, height

# Traiter chaque image et sa correspondance XML
for image_file in image_files:
    xml_file = image_file.rsplit(".", 1)[0] + ".xml"
    xml_path = os.path.join(train_images_path, xml_file)

    # V√©rifier si le fichier XML existe
    if os.path.exists(xml_path):
        tree = ET.parse(xml_path)
        root = tree.getroot()

        # Charger les dimensions de l'image
        img_path = os.path.join(train_images_path, image_file)
        img = cv2.imread(img_path)
        img_height, img_width, _ = img.shape

        # Cr√©er le fichier d'annotations pour cette image
        annotation_file = os.path.join(output_annotations_path, image_file.rsplit(".", 1)[0] + ".txt")
        with open(annotation_file, "w") as f:
            for obj in root.findall("object"):
                name = obj.find("name").text
                bbox = obj.find("bndbox")
                xmin = int(bbox.find("xmin").text)
                ymin = int(bbox.find("ymin").text)
                xmax = int(bbox.find("xmax").text)
                ymax = int(bbox.find("ymax").text)

                # Normaliser les coordonn√©es
                center_x, center_y, width, height = normalize_coordinates(xmin, ymin, xmax, ymax, img_width, img_height)

                # Enregistrer l'annotation au format YOLO
                # Ici, 0 est l'ID de la classe (car c'est la plaque d'immatriculation)
                f.write(f"0 {center_x} {center_y} {width} {height}\n")

        print(f"‚úÖ Annotations g√©n√©r√©es pour {image_file}")
    else:
        print(f"‚ùå Pas de fichier XML trouv√© pour {image_file}")

print("‚úÖ Toutes les annotations ont √©t√© g√©n√©r√©es avec succ√®s !")
````

````python
import os
import random
import shutil

# Dossier contenant les images et annotations d'entra√Ænement
train_images_path = "/content/dataset/dataset/train/"
train_labels_path = "/content/dataset/dataset/train/"

# Cr√©er les dossiers pour la validation
valid_images_path = "/content/dataset/valid/images"
valid_labels_path = "/content/dataset/valid/labels"

os.makedirs(valid_images_path, exist_ok=True)
os.makedirs(valid_labels_path, exist_ok=True)

# Lister les images d'entra√Ænement
all_images = [f for f in os.listdir(train_images_path) if f.endswith(('.jpg', '.jpeg', '.png'))]
print(all_images)

# S√©lectionner 20% des images pour la validation
valid_images = random.sample(all_images, int(0.2 * len(all_images)))

# D√©placer les images et leurs annotations correspondantes vers le dossier de validation
for image in valid_images:
    # D√©placer l'image
    shutil.move(os.path.join(train_images_path, image), os.path.join(valid_images_path, image))

    # D√©placer l'annotation correspondante
    label_file = image.rsplit('.', 1)[0] + '.xml'
    label_file_path = os.path.join(train_labels_path, label_file)

    if os.path.exists(label_file_path):
        shutil.move(label_file_path, os.path.join(valid_labels_path, label_file))
    else:
        print(f"‚ö†Ô∏è Annotation file {label_file} not found for image {image}")

print(f"‚úÖ {len(valid_images)} images ont √©t√© d√©plac√©es vers le dossier de validation.")
````

````python
import os
import random
import shutil
from xml.etree import ElementTree as ET

# Function to convert XML annotations to YOLO format
def convert_xml_to_yolo(xml_path, output_path, class_id=0):
    tree = ET.parse(xml_path)
    root = tree.getroot()

    # Get image dimensions
    size = root.find("size")
    width = int(size.find("width").text)
    height = int(size.find("height").text)

    # Parse bounding boxes
    yolo_annotations = []
    for obj in root.findall("object"):
        bndbox = obj.find("bndbox")
        xmin = int(bndbox.find("xmin").text)
        ymin = int(bndbox.find("ymin").text)
        xmax = int(bndbox.find("xmax").text)
        ymax = int(bndbox.find("ymax").text)

        # Convert to YOLO format (normalized center x, center y, width, height)
        x_center = (xmin + xmax) / 2 / width
        y_center = (ymin + ymax) / 2 / height
        box_width = (xmax - xmin) / width
        box_height = (ymax - ymin) / height

        yolo_annotations.append(f"{class_id} {x_center} {y_center} {box_width} {box_height}")

    # Write YOLO annotations to .txt file
    with open(output_path, "w") as f:
        f.write("\n".join(yolo_annotations))

# Base dataset path
base_dataset_path = "/content/dataset/dataset"
new_dataset_path = "/content/dataset"

# Create new directory structure
for folder in ["train", "valid", "test"]:
    os.makedirs(os.path.join(new_dataset_path, folder, "images"), exist_ok=True)
    os.makedirs(os.path.join(new_dataset_path, folder, "labels"), exist_ok=True)

# Function to move and convert files
def organize_dataset(source_images_path, source_labels_path, dest_images_path, dest_labels_path, split_ratio=1.0):
    # List all images
    all_images = [f for f in os.listdir(source_images_path) if f.endswith(('.jpg', '.jpeg', '.png'))]
    random.shuffle(all_images)

    # Select images based on split ratio
    split_index = int(len(all_images) * split_ratio)
    selected_images = all_images[:split_index]

    # Move images and convert annotations
    for image in selected_images:
        # Move image
        shutil.move(os.path.join(source_images_path, image), os.path.join(dest_images_path, image))

        # Convert and move annotation
        xml_file = image.rsplit('.', 1)[0] + '.xml'
        txt_file = image.rsplit('.', 1)[0] + '.txt'
        xml_path = os.path.join(source_labels_path, xml_file)
        txt_path = os.path.join(dest_labels_path, txt_file)

        if os.path.exists(xml_path):
            convert_xml_to_yolo(xml_path, txt_path)
            os.remove(xml_path)  # Remove the original XML file
        else:
            print(f"‚ö†Ô∏è Annotation file {xml_file} not found for image {image}")

# Organize train and valid datasets
organize_dataset(
    source_images_path=os.path.join(base_dataset_path, "train"),
    source_labels_path=os.path.join(base_dataset_path, "train"),
    dest_images_path=os.path.join(new_dataset_path, "train", "images"),
    dest_labels_path=os.path.join(new_dataset_path, "train", "labels"),
    split_ratio=0.8  # 80% for training
)

organize_dataset(
    source_images_path=os.path.join(base_dataset_path, "train"),
    source_labels_path=os.path.join(base_dataset_path, "train"),
    dest_images_path=os.path.join(new_dataset_path, "valid", "images"),
    dest_labels_path=os.path.join(new_dataset_path, "valid", "labels"),
    split_ratio=0.2  # 20% for validation
)

# Organize test dataset
organize_dataset(
    source_images_path=os.path.join(base_dataset_path, "test"),
    source_labels_path=os.path.join(base_dataset_path, "test"),
    dest_images_path=os.path.join(new_dataset_path, "test", "images"),
    dest_labels_path=os.path.join(new_dataset_path, "test", "labels"),
    split_ratio=1.0  # 100% for testing
)

print("‚úÖ Dataset reorganization and annotation conversion completed.")
````

````python
from ultralytics import YOLO
import cv2
from PIL import Image
import IPython.display as display
# Load the trained YOLOv8 model
model = YOLO("/content/best.onnx")  # Replace with your trained model path
# Load the image
image_path = "/content/renault-symbol-tunisia-2.jpg"
image = cv2.imread(image_path)
# Run inference
results = model(image)
# Extract bounding box coordinates
for result in results:
    for box in result.boxes.xyxy:  # Get (x1, y1, x2, y2) bounding box coordinates
        x1, y1, x2, y2 = map(int, box)  # Convert to integers
        # Crop the detected license plate
        license_plate = image[y1:y2, x1:x2]
        # Save the cropped image
        cropped_path = "cropped_plate.jpg"
        cv2.imwrite(cropped_path, license_plate)
        print(f"License plate saved at {cropped_path}")
# Display the cropped image
cropped_image = Image.open("cropped_plate.jpg")
display.display(cropped_image)
````

````python
import cv2
import os
from ultralytics import YOLO

# Load the trained YOLO model
model = YOLO("/content/best.pt")  # Replace with your trained model

# Load the video
video_path = "/content/WhatsApp Video 2025-02-07 at 13.24.05_c53b8edc.mp4"  # Replace with your video file path
cap = cv2.VideoCapture(video_path)

# Get video properties
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define output video file
output_video_path = "output_video.mp4"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

# Create a folder to save cropped plates
if not os.path.exists("cropped_plates"):
    os.makedirs("cropped_plates")

frame_count = 0  # To track frames

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Stop if no frame is left

    # Run YOLOv8 detection
    results = model(frame)

    for i, result in enumerate(results):
        for j, box in enumerate(result.boxes.xyxy):
            x1, y1, x2, y2 = map(int, box)  # Get bounding box coordinates

            # Crop and save the license plate
            license_plate = frame[y1:y2, x1:x2]
            cropped_plate_path = f"cropped_plates/plate_{frame_count}_{j}.jpg"
            cv2.imwrite(cropped_plate_path, license_plate)

            # Draw bounding box on frame
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # Write frame to output video
    out.write(frame)
    frame_count += 1

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Processed video saved at: {output_video_path}")
print(f"Cropped license plates saved in 'cropped_plates/' folder")
````
